```python
#Q5
cust=['CUST_ID']
x=product.groupby(cust).agg({'PURCHASE_VALUE':'sum'}).reset_index()
max_cust=x.PURCHASE_VALUE.max()


print(x.loc[x.PURCHASE_VALUE==max_cust])
```

```python
df = pd.merge(customer, product, on=cust, how='right')
```

```python

df['row_num'] = df.reset_index().index
df.head()
```

```python
#Q6
col=['CUST_CITY']
y=df.groupby(col).count().reset_index()
max_quant=y['row_num'].max()

print(y.loc[y.row_num==max_quant]['CUST_CITY'])
```
